### Load

```python
from sklearn.datasets import fetch_20newsgroups

categories = [
    'alt.atheism',
    'talk.religion.misc',
    'comp.graphics',
    'sci.space',
]

dataset = fetch_20newsgroups(subset='all', 
                             shuffle=True, random_state=42)

print("%d documents" % len(dataset.data))
print("%d categories" % len(dataset.target_names))
print()

labels = dataset.target
labels_names = dataset.target_names
print(labels)
true_k = np.unique(labels).shape[0]
```

### tf-idf vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_df=0.5, decode_error="ignore", stop_words='english', min_df=3)
X = vectorizer.fit_transform(dataset.data)
```

```python
from sklearn.decomposition import TruncatedSVD
lsa = TruncatedSVD()
X_truncated = lsa.fit_transform(X)
print(X.shape)
print(X_truncated.shape)
```

### Potting function

```python
from itertools import cycle
import pylab as pl
def plot_2d(X, y, labels):
    colors = cycle("rgbcykw")
    pl.figure(figsize=(12, 10), dpi=80)
    labels_ids = range(len(labels))
    for i, c, l in zip(labels_ids, colors, labels):
        pl.scatter(X[y == i, 0], X[y == i, 1], c=c, label=l)
    
    pl.legend()
```

### Plotting original labels

```python
plot_2d(X_truncated, labels, labels_names)
```

# cluster with original data

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
km.fit(X)
```

```python
plot_2d(X_truncated, km.labels_, ['cluster %s' % i for i in range(20)])
```

```python
from sklearn import metrics
from time import time

t0 = time()
km.fit(X)
print("done in %0.3fs" % (time() - t0))

print("Homogeneity: %0.3f" % metrics.homogeneity_score(labels, km.labels_))
print("Completeness: %0.3f" % metrics.completeness_score(labels, km.labels_))
print("V-measure: %0.3f" % metrics.v_measure_score(labels, km.labels_))
print("Adjusted Rand-Index: %.3f"
      % metrics.adjusted_rand_score(labels, km.labels_))
print("Silhouette Coefficient: %0.3f"
      % metrics.silhouette_score(X, labels, sample_size=1000))
```
